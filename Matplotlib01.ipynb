### How to show plot
1. Script
- Script에서는 plt.show()로 호출해야지 plot이 그려진다
- plt.show()는 현재 활성화 되어있는 모든 그림 객체를 찾아서 그림을 표시하는 하나 이상의 대화창을 연다
- 하나 이상의 대화창에 대한 예시는 Matplotlib_test/script01.py 실행해보자
- plt.show() script 혹은 파이썬 세션당 한번 불러야지 안그러면 백엔드 종속적인 동작 이르킨다. 이에 대한 예시는 atplotlib_test/script02.py 참고
2. Notebook
- %matplotlib inline -> default는 이거다
- %matplotlib notebook

Defulat로 inline이 설정되어있기 때문에 굳이 show를 호출하지 않아도 그림이 cell밑에 그려진다

### matplotlib 은 두가지의 인터페이스를 제공한다
- matlab style 
    - 가장 익숙한 방식이다
- objective oriented
    - 주의해야할 점은 plt에서 지원했던 형식이랑 조금 다른부분이 분명히 있다

#### 1. matlab style

```python
# matlab style
x = np.linspace(1,10,100)
plt.plot(x,np.sin(x)) # show 없어도 문제없다
```

```python
# matlab style
# multiple plot
plt.subplot(2,1,1) # 2행 1열 중 첫번째꺼
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.sin(x))

```

```python
# 크기가 안맞으면 어케 될까? => 한 figure내에서는 이런 바보같은 방식으로는 불가능하다
plt.subplot(2,1,1)      # 출력 안될거다
plt.plot(x,np.sin(x),color='green')

plt.subplot(3,1,2)
plt.plot(x,np.sin(x),color='red')
plt.subplot(3,1,1)
plt.plot(x,np.sin(x),color='brown')
```

```python
# 위에 문제를 해결하고 싶으면 객체로 figure를 각각 만들어라 
plt.figure()
plt.subplot(2,1,1)      
plt.plot(x,np.sin(x),color='green')

plt.figure()
plt.subplot(3,1,2)
plt.plot(x,np.sin(x),color='red')
plt.subplot(3,1,1)
plt.plot(x,np.sin(x),color='brown')
```

#### figure의 역할은 뭘까?

```python
# inline이기 때문에 여기서는 굳이 plt.figure()를 불러주지 않아도 문제가 없었다
x = np.linspace(1,10,100)
plt.plot(x,np.sin(x))
```

```python
# figure은 그래프의 전체적인 틀을 담는 container다
plt.figure(figsize=(4,4)) 
plt.axes()
```

#### 2. object oriented style

```python
# figure 하나를 처리하는 
fig = plt.figure()
ax = plt.axes()
x = np.linspace(1,10,100)
ax.plot(x,np.sin(x))

# type을 한번씩 봐주자
print(type(fig)) 
print(type(ax))
```

```python
# fig.add_subplot 활용
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
x = np.linspace(1,10,100)
ax1.plot(x,np.sin(x))
ax2.plot(x,np.sin(x),color='red')
```

```python
# 사이즈 안맞으면 아래같이 이상하게 figure가 나온다 유의하자
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(3,1,2)
x = np.linspace(1,10,100)
ax1.plot(x,np.sin(x))
ax2.plot(x,np.sin(x),color='red')
```

#### plt.subplots()
- 객체 지향 인터페이스에서 가장 많이 사용되는 형태일거다

```python
# plt.subplots()
fig, ax = plt.subplots() # defualt is nrow=1, ncol=1
x = np.linspace(1,10,100)
ax.plot(x,np.sin(x))

# type을 한번씩 봐주자
print(type(fig)) 
print(type(ax))
```

```python
# plt.subplots()
fig, ax = plt.subplots(1,1) # 행,열
x = np.linspace(1,10,100)
ax.plot(x,np.sin(x))

# type을 한번씩 봐주자
print(type(fig)) 
print(type(ax))
```

```python
# 여러개의 subplot은?

# plt.subplots()
fig, ax = plt.subplots(4) 
x = np.linspace(1,10,100)
ax[0].plot(x,np.sin(x))

# type을 한번씩 봐주자
print(type(fig)) 
print(type(ax)) # 타입을 보면 갑자기 ndarray로 바뀌어있다
print(ax) # 4개의 axes들이 다 담겨있는 형태를 확인할수 있다
```

```python
# 디스플레이 되는 순서도 물론 바꿀수있다
# => 행열의 컨셉 이용

# plt.subplots()
fig, ax = plt.subplots(1,2) 
x = np.linspace(1,10,100)
ax[0].plot(x,np.sin(x))

```

```python
fig, ax = plt.subplots(2,2) 
x = np.linspace(1,10,100)
print(ax) # ax의 출력물을 잘보면 1차원의 nd array가 아닌 input으로 받은 행렬의 형태 에 따른 ndarray가 들어가 있다
ax[0,1].plot(x,np.sin(x)) #1행 2열
```

#### tuple unpack활용

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(4) # 괄호에 주의하자
x = np.linspace(1,10,100)
ax1.plot(x,np.sin(x))
ax2.plot(x,np.sin(x),color='orange')
ax3.plot(x,np.sin(x),color='green')
ax4.plot(x,np.sin(x),color='red')
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2) # 괄호에 주의하자
x = np.linspace(1,10,100)
ax1.plot(x,np.sin(x))
ax2.plot(x,np.sin(x),color='orange')
ax3.plot(x,np.sin(x),color='green')
ax4.plot(x,np.sin(x),color='red')
```
