```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Figure 와 Axes
*인터페이스를 이해하기 전에 matplotlib의 plot이 그려지는 방식을 이해할 필요가 있다*

**Figure**

plt.figure() 인스턴스는 축과 그래픽, 텍스트, 레이블을 표시하는 모든 객체를 포함하는 하나의 컨테이너다 

**Axes**

plt.axes() 인스터스는 축을 의미한다 => 실질적인 plotting은 축에서 이루어지는 것이다 

### matplotlib 은 두가지의 인터페이스를 제공한다
- matlab style 
    - 가장 중요한 부분은 matlab style은 인터페이스 자체가 상태를 저장한다
    - 이로 인해서 Figure axes 등을 객체로 저장할 필요가 없다
- objective oriented
    - 간단한 그래프는 matlab style에서 빠르게 처리가 가능하나 복잡한 그래프에서는 객체 style이 도움이 된다 
    - ex) 첫번째 패널 생성 후 두번째 패널 작업 중에 첫번째 패널 수정이 matlab style에서는 불편하다

#### 1. matlab style

```python
# matlab style
x = np.linspace(1,10,100)
plt.figure()
plt.plot(x,np.sin(x))
plt.show()
```

```python
# Ipython cell 에서는 plt.figure()를 부르지도 않아도 가능하다
# 자동으로 생성되서 저장하는거 같다

print(plt.gcf()) # plt.figure() 부르지도 않았는데 자동으로 생성되 있다
plt.show() # axes는 아직 생성 안되서 그림이 그려지진 않는거 같다
```

```python
# Ipython cell 에서는 plt.figure()를 부르지도 않아도 가능하다
# axes도 자동으로 부를때 생성되있다
print(plt.gca()) 
plt.show()
```

```python
# matlab style multiple plot

# plt.figure()
plt.subplot(2,1,1) # 2행 1열 중 첫번째 패널 생성하고 현재 축으로 지정
plt.plot(x,np.sin(x)) # 지정된 축에 plot그림
plt.subplot(2,1,2) #  2행 1열 중 두번재 패널 생성하고 현재 축으로 지정
plt.plot(x,np.sin(x)) # 지정된 축에 plot그림

```

```python
type(plt.subplot(2,1,1)) # 이 자체가 사실 axes 객체다
```

```python
# 크기가 안맞으면 어케 될까? => 한 figure내에서는 이런 바보같은 방식으로는 불가능하다
plt.subplot(2,1,1)      # 출력 안될거다
plt.plot(x,np.sin(x),color='green')

plt.subplot(3,1,2)
plt.plot(x,np.sin(x),color='red')
plt.subplot(3,1,1)
plt.plot(x,np.sin(x),color='brown')
```

```python
# 위에 문제를 해결하고 싶으면 객체로 figure를 각각 만들어라 

plt.figure(figsize=(6,4)) # figure 생성하고 현재 figure로 지정
print(plt.gcf())
plt.subplot(2,1,1)      
plt.plot(x,np.sin(x),color='green')
plt.show()

 # figure 생성하고 현재 figure로 지정
plt.figure(figsize=(6,5)) 
print(plt.gcf()) # 위랑 다른 figure 불린다
plt.subplot(3,1,2)
plt.plot(x,np.sin(x),color='red')
plt.subplot(3,1,1)
plt.plot(x,np.sin(x),color='brown')
plt.show()
```

#### 2. object oriented style

```python
# figure 하나를 처리하는 
fig = plt.figure()
ax = plt.axes()
x = np.linspace(1,10,100)
ax.plot(x,np.sin(x)) # 축에다가 그린다

# type을 한번씩 봐주자
print(type(fig)) 
print(type(ax))
```

```python
# 위에서 보았던 matlab style도 그래도 object style로 사용 가능
fig = plt.figure() 
ax = plt.subplot(2,1,1) # 축 생성 저장 
ax.plot(x,np.sin(x))
ax = plt.subplot(2,1,2) # 축 생성 저장
ax.plot(x,np.cos(x))
```

```python
# fig.add_subplot 활용
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1) # figure에 축  추가 
ax2 = fig.add_subplot(2,1,2)
x = np.linspace(1,10,100)
ax1.plot(x,np.sin(x))
ax2.plot(x,np.sin(x),color='red')
```

```python
# 사이즈 안맞으면 아래같이 이상하게 figure가 나온다 유의하자
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(3,1,2)
x = np.linspace(1,10,100)
ax1.plot(x,np.sin(x))
ax2.plot(x,np.sin(x),color='red')
```

#### plt.subplots()
- 객체 지향 인터페이스에서 가장 많이 사용되는 형태일거다

```python
plt.subplots(2) # 2행 1 열
# 출력을 잘 보면
# fig class 1개
# axes class 2개 돌려준다
```

```python
# plt.subplots()
fig, ax = plt.subplots() # defualt is nrow=1, ncol=1
x = np.linspace(1,10,100)
ax.plot(x,np.sin(x))
```

```python
# plt.subplots()
fig, ax = plt.subplots(1,1) # 행,열
x = np.linspace(1,10,100)
ax.plot(x,np.sin(x))

```

```python
# 여러개의 subplot은?

# plt.subplots()
fig, ax = plt.subplots(4) # 4행 1열
x = np.linspace(1,10,100)
ax[0].plot(x,np.sin(x))

# type을 한번씩 봐주자
print(type(fig)) 
print(type(ax)) # 타입을 보면 갑자기 ndarray로 바뀌어있다
print(ax) # 4개의 axes들이 다 담겨있는 형태를 확인할수 있다
```

```python
fig, ax = plt.subplots(1,4) # 1 행 2열
x = np.linspace(1,10,100)
ax[0].plot(x,np.sin(x)) 
ax[2].plot(x,np.cos(x))

```

```python
fig, ax = plt.subplots(2,2) 
x = np.linspace(1,10,100)
print(ax) # ax의 출력물을 잘보면 1차원의 nd array가 아닌 input으로 받은 행렬의 형태 에 따른 ndarray가 들어가 있다
ax[0,1].plot(x,np.sin(x)) #1행 2열
```

#### tuple unpack활용

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(4) # 4*1
x = np.linspace(1,10,100)
ax1.plot(x,np.sin(x))
ax2.plot(x,np.sin(x),color='orange')
ax3.plot(x,np.sin(x),color='green')
ax4.plot(x,np.sin(x),color='red')
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2) # 괄호에 주의하자

# ax1,ax2 1행
# ax3,ax4 2행
# ax1 -> 1행 1열
# ax2 -> 1행 2열

x = np.linspace(1,10,100)
ax1.plot(x,np.sin(x))
ax2.plot(x,np.sin(x),color='orange')
ax3.plot(x,np.sin(x),color='green')
ax4.plot(x,np.sin(x),color='red')
```
