```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# style available
plt.style.available
```

```python
# style set
plt.style.use('seaborn-whitegrid')
```

### 선 그래프

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
```

### 선 색상과 스타일
- 아래 예시 말고도 지원하는게 더 있다 근데 그렇게 쓸거 같지는 않다 

```python
x = np.linspace(0,10,1000)

plt.plot(x,np.sin(x-0),color='blue') # 색상 이름
plt.plot(x,np.sin(x-1),color='g')    # 짧은 색상 코드
plt.plot(x,np.sin(x-2),color='0.75') # 0-1 사이로 회색조
plt.plot(x,np.sin(x-3),color='chartreuse') # 모든 html 색상 이름 지원
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,x+0,linestyle = 'solid')
plt.plot(x,x+1,linestyle = 'dashed') # 자동으로 축에 한번 더 그리면 색깔을 바꿔주기도 한다
plt.plot(x,x+2,linestyle = 'dashdot')
plt.plot(x,x+3,linestyle = 'dotted')
```

```python
# linestyle keyword로 가능
x = np.linspace(0,10,1000)
plt.plot(x,x+0,linestyle = '-')
plt.plot(x,x+1,linestyle = '--') # 자동으로 축에 한번 더 그리면 색깔을 바꿔주기도 한다
plt.plot(x,x+2,linestyle = '-.')
plt.plot(x,x+3,linestyle = ':')
```

```python
# linestyle keywword + color code
x = np.linspace(0,10,1000)
plt.plot(x,x+0,'-g')
plt.plot(x,x+1,'--c') 
plt.plot(x,x+2,'-.k')
plt.plot(x,x+3,':r')
```

### 축 경계

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5]) # [xlim,xlim,ylim,ylim]
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
# 가로 세로 비율 균등하게 설정
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.axis('equal')
```

### 제목, 레이블, 간단한 범례

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('A sin curve')
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.legend()
```

### object style 적용

```python
x = np.linspace(0,10,1000)
fig,ax = plt.subplots()
ax.plot(x,np.sin(x))
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.set_xlim(-1,11)
ax.set_ylim(-1.5,1.5)
ax.set_title('A sin curve')
```

```python
# objec style은 set으로 한번에 지정 가능

x = np.linspace(0,10,1000)
fig,ax = plt.subplots()
ax.plot(x,np.sin(x))
ax.set(
    xlim=(-1,11),ylim=(-1.5,1.5),
    xlabel='x',ylabel='sin(x)',
    title='A sin curve'
)
```

```python
x = np.linspace(0,10,1000)
fig,ax = plt.subplots()
ax.plot(x,np.sin(x),'-g',label='sin(x)')
ax.plot(x,np.cos(x),':b',label='cos(x)')
ax.legend()
```
