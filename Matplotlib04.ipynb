```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

### 산점도
1. plt.plot + marker
2. plt.scatter

#### 1. plt.plot 이용한 산점도

```python
x= np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,marker='o',linestyle='')
```

```python
x= np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'o') # 위에 보면 marker + linestyle 합쳐서 인식한다
```

```python
x= np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'og') # 위에 보면 marker + linestyle 합쳐서 인식한다
```

```python
x= np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'--og') # marker ='o', linestyle = '--' , color='g' 가 합쳐진 형태
```

```python
# 여러가지 마커 종류 존재
markers = ['o','.',',','x','s','d']
for m in markers:
    plt.plot(np.random.rand(5),np.random.rand(5),m,label=f"marker ='{m}'")
plt.legend(numpoints=1)
```

```python
# marker 모양 말고 더 세부적인 속성도 존재한다
x= np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'o',markerfacecolor='white',markeredgecolor='g',markeredgewidth=2) 
```

#### 2.plt.scatter 이용한 산점도
- plt.plot과의 가장 큰 차이점은 각 점의 속성을 개별적으로 제어 가능하다는 점이다 => 데이터가 너무 많을때는 plt.scatter하면 오히려 비효율적일수 있다
- plt.plot에서는 다 모양이 동일한 산점도였다


```python
x= np.linspace(0,10,30)
y = np.sin(x)
plt.scatter(x,y,'o') # 에러난다 plt.scatter에서는 quick 지원안한다
```

```python
x= np.linspace(0,10,30)
y = np.sin(x)
plt.scatter(x,y,marker='o')
```

```python
# size, color, 각각 설정 가능하다
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000*np.random.randn(100)

plt.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.3)
plt.colorbar()

# color map 'viridis' 통해서 randn 값들이 mapping됬다
```

```python
# 실제 데이터 사용
from sklearn.datasets import load_iris
iris = load_iris()
import pandas as pd
iris_df = pd.DataFrame(iris.data)
iris_df.columns = iris.feature_names
iris_df
```

```python
feature1 = iris_df.iloc[:,0]
feature2 = iris_df.iloc[:,1]
feature3 = iris_df.iloc[:,2]
plt.scatter(feature1,feature2,s=100*feature3,c=iris.target,cmap='viridis',alpha=0.3)

# 4가지 차원 동시 파악 가능
# featuer 1+2+3 에다가 종류가 색깔로 들엉가있다
```
