### 오차 시각화 하기 

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy,fmt='-..g') # 동일한 축약 기호사용 가능
```

```python
# errorbar 자체도 조정가능하다

x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy,fmt='-..g',ecolor='gray',elinewidth=2,)
```

```python
# error 크기를 각각 인식하기도 한다
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)
plt.errorbar(x,y,yerr=np.linspace(0,1,50),fmt='-..g',ecolor='gray',elinewidth=2,)
```

### 연속 오차
- 연속 오차에서는 막대 오차 그래프는 말이 안된다
- fill_between 사용하자
- fill_between(x,y하한,y상한)

```python
x = np.linspace(0,10,100)
plt.plot(x,2*x,'-')
plt.plot(x,2*x-2,'-')
plt.fill_between(x,2*x-2,2*x,color='gray',alpha=0.3)
```

```python
x = np.linspace(0,10,100)
plt.plot(x,x**2,'-')
plt.fill_between(x,0,x**2,color='skyblue')
```

```python
x = np.linspace(0,10,100)
plt.plot(x,x**2,'-')
plt.fill_between(x,x**2,max(x**2),color='skyblue')
```

```python
x = np.linspace(0,10,100)
plt.plot(x,x**2,'-')
plt.fill_between(x,2,4,color='skyblue')
```

```python
x = np.linspace(0,10,100)
plt.plot(x,x**2,'-')
plt.fill_betweenx(x**2,2,4,color='skyblue') # y, x하한,x상한
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'.r')
plt.plot(x,np.sin(x),'-',color='gray',alpha=0.5)

plt.fill_between(x,np.sin(x)-1,np.sin(x)+1,color='gray',alpha=0.2)
```

### Histogram

```python
# color option
x = np.random.randn(1000)
plt.hist(x,color='g',edgecolor='b')
```

```python
# bins
x = np.random.randn(1000)
fig,axes = plt.subplots(1,2)
axes[0].hist(x,bins=30)
axes[1].hist(x,bins=10)
plt.show()
```

```python
# bins 내맘대로 정할수도 있다
x = np.random.randn(1000)
bins = np.linspace(-3,3,100)
plt.hist(x,bins=bins)
plt.show()
```

```python
# density => 그래프 밀도로 보여준다
x = np.random.randn(1000)
plt.hist(x,density=True,color='g',edgecolor='b')
```

```python
# histtype
mu_x = 200
sigma_x = 25
x = np.random.normal(mu_x, sigma_x, size=100)

mu_w = 200
sigma_w = 10
w = np.random.normal(mu_w, sigma_w, size=100)

fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].hist(x, 20, histtype='stepfilled', facecolor='g',
               alpha=0.75)
axs[0, 0].set_title('stepfilled')

axs[0, 1].hist(x, 20, histtype='step', facecolor='g',
               alpha=0.75)
axs[0, 1].set_title('step')

axs[1, 0].hist([x,w], histtype='barstacked', rwidth=0.8) # multiple data => stacked
axs[1, 0].set_title('barstacked')

# Create a histogram by providing the bin edges (unequally spaced).
bins = [100, 150, 180, 195, 205, 220, 250, 300]
axs[1, 1].hist([x,w], bins, histtype='bar', rwidth=0.8) # multiple data => arranged side by side
axs[1, 1].set_title('bar, unequal bins')

fig.tight_layout()
plt.show()
```

```python
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)


fig, axs = plt.subplots(nrows=2, ncols=2)
axs[0,0].hist([x1,x2,x3]) # defualt는 bar이기 때문에 옆으로 arrange됨
axs[0,1].hist([x1,x2,x3],histtype='barstacked',alpha=0.5) # 각 구간별 각각의 값을 보여준다 => 높이는 합산한거다
axs[1,0].hist([x1,x2,x3],histtype='step')
axs[1,1].hist([x1,x2,x3],histtype='stepfilled',alpha=0.5)
plt.show()
```

```python
### 2차원 histogram
mean = [0,0]
cov = [ [1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,10000).T

plt.hist2d(x,y,30,cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```
