```python
import matplotlib.pyplot as plt
import numpy as np
```

### two y-axis

```python
# 범위를 동일하게 지정해주지 않으면 아래처럼 막무가내로 표시된다 
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
ax1.plot(x, y1, color='green')

ax2 = ax1.twinx()
ax2.set_ylabel('2nd Y-Axis') 
ax2.plot(x, y2, color='deeppink')

plt.show()
```

```python
# 범위를 동일하게 지정해줫다
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
ax1.set_ylim(0,3)
ax1.plot(x, y1, color='green')

ax2 = ax1.twinx()
ax2.set_ylabel('2nd Y-Axis') 
ax2.set_ylim(-5,5)
ax2.plot(x, y2, color='deeppink')

plt.show()
```

```python
# 범례 표시
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
ax1.set_ylim(0,3)
ax1.plot(x, y1, color='green',label='line1')

ax2 = ax1.twinx()
ax2.set_ylabel('2nd Y-Axis') 
ax2.set_ylim(-5,5)
ax2.plot(x, y2, color='deeppink',label='line2')

ax1.legend()
ax2.legend()
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
ax1.set_ylim(0,3)
line1 = ax1.plot(x, y1, color='green')

ax2 = ax1.twinx()
ax2.set_ylabel('2nd Y-Axis') 
ax2.set_ylim(-5,5)
line2 = ax2.plot(x, y2, color='deeppink')

from matplotlib.legend import Legend
leg = Legend(ax1,line1+line2,['line1','line2'],loc='upper left')
ax1.add_artist(leg)

# 이게 더하는게 다른 타입의 그림에서는 안될수도 있다 그럴땐 각자 부르자

plt.show()
```

### two type grphs

```python
from matplotlib.legend import Legend

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])

# 3. 그래프 그리기
fig, ax1 = plt.subplots()

line = ax1.plot(x, y1, '-s', color='green', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax1.set_ylim(0, 18)
ax1.set_xlabel('Year')
ax1.set_ylabel('Price ($)')
ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx()
bar = ax2.bar(x, y2, color='deeppink', label='Demand', alpha=0.7, width=0.7)
ax2.set_ylim(0, 18) # twin x한다고 해서 ylim까지 그대로 copy해오진 않는다 이것도 설정해줘야함
ax2.set_ylabel(r'Demand ($\times10^6$)')
ax2.tick_params(axis='y', direction='in')


# 다른 축이기때문에 따로 불러준다
ax1.legend()
ax2.legend(loc='upper left')

plt.show()

# 우선 ax1.twinx()로 X축을 공유하는 이중 Y축을 만들고,

# ax1.plot()과 ax2.bar()를 사용해서 y1, y2 데이터를 각각 꺾은선 그래프와 막대 그래프의 형태로 나타냈습니다.
```
