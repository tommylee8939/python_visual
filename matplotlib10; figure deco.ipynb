```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
import pandas as pd
```

```python
x = np.random.randn(1000)
plt.hist(x)
```

```python

plt.figure(facecolor='None') # 그래프 배경 없애기
ax = plt.axes(facecolor='#E6E6E6') # plot 배경
ax.set_axisbelow(True)

# grid
ax.grid(color='w',linestyle='solid')

# 위쪽 오른쪽 눈금 숨김
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# 축 선 숨김 (바깥 축선)
for spine in ax.spines.values():
    spine.set_visible(False)

x = np.random.randn(1000)
ax.hist(x,edgecolor='#E6E6E6',color='orange') 
```

```python
# subplot 간격
fig,ax = plt.subplots(2,3)
fig.subplots_adjust(hspace=10,wspace=10)
```
